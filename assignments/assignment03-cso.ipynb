This program retrieves the dataset for the "exchequer account (historical series)" from the CSO, and stores it into a file called "cso.json".

```python
# import required packages
import requests
import json
```

requests → gets data from a website/API

json → saves the data as a .json file

```python
def get_cso_data():
    url = "https://ws.cso.ie/public/api.restful/PxStat.Data.Cube_API.ReadDataset/FIQ02/JSON-stat/2.0/en"
    response = requests.get(url)
    data = response.json()

    with open("cso.json", "w") as file:
        json.dump(data, file)
```

```python
print("Dataset saved to cso.json")
```
